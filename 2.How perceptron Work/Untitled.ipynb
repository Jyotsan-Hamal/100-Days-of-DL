```python
import numpy as np
import seaborn as sns
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
x,y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,random_state=41,n_clusters_per_class=1,n_classes=2,hypercube=False,class_sep=10)
```

```python
x
```

```python
sns.scatterplot(x[:,0],x[:,1],hue=y)
```

```python
def perceptron(x,y):
    x=np.insert(x,0,1,axis=1)
    weights = np.ones(x.shape[1])
    lr=0.01
    for i in range(5000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(weights,x[j]))
        weights = weights + lr*(y[j]-y_hat)*x[j]
    return weights[0],weights[1:]
        
```

```python
def step(x):
    if x>0:
        return 1
    elif x<0:
        return 0
    
```

```python
y_intercept,coef_=perceptron(x,y)
```

```python
coef_
```

```python
y_intercept
```

```python
m = -(coef_[0]/coef_[1])
b = -(y_intercept/coef_[1])
```

```python
line = m*x[0]
```

```python
x_input = np.linspace(3,-3,100)
y_input = m*x_input+b
```
